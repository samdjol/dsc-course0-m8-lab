```python
# loading relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis  
- Load in the cleaned data

```python
df = pd.read_csv('data/CleanedAviationData.csv')
df.head()
```

## Explore safety metrics across models/makes
- Remember that the client is interested in separate recommendations for smaller airplanes and larger airplanes. Choose a passenger threshold of 20 and separate the plane types. 

```python
df.loc[df['Total Passangers'] < 20, 'Size'] = 'Small'
df['Size'] = df['Size'].fillna('Large')
```

```python
df.head()
```

#### Analyzing Makes

Explore the human injury risk profile for small and larger Makes:
- choose the 15 makes for each group possessing the lowest mean fatal/seriously injured fraction
- plot the mean fatal/seriously injured fraction for each of these subgroups side-by-side

```python
small_plane = df[df['Size'] == 'Small']
large_plane = df[df['Size'] == 'Large']

#After dividing the dataframes, we run the risk again of working insufficient sample sizes. 

#Put the 50 makes threshold on the small plane dataframe. The large dataframe is much smaller. Threshold set to 20.

small_counts = small_plane['Make'].value_counts()

small_plane = small_plane[small_plane['Make'].isin(small_counts[small_counts >= 50].index)]

large_counts = large_plane['Make'].value_counts()

large_plane = large_plane[large_plane['Make'].isin(large_counts[large_counts >= 20].index)]

large_plane['Make'].value_counts()
```

```python
#Grouping by make, we can pull the lowest mean fatal/serious injury fraction for each category.
small_plane_chance = small_plane.groupby('Make')['Fatal/Serious Injury Chance'].mean()
large_plane_chance = large_plane.groupby('Make')['Fatal/Serious Injury Chance'].mean()
```

```python
#Here we are getting the lowest 15 makes by injury fraction mean for small and large
small_15 = small_plane_chance.sort_values().head(15).reset_index()
small_15.columns = ['Make', 'Fatal/Serious Injury Chance']
small_15['Type'] = 'Small'

large_15 = large_plane_chance.sort_values().head(15).reset_index()
large_15.columns = ['Make', 'Fatal/Serious Injury Chance']
large_15['Type'] = 'Large'
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(data=small_15, x='Make', y='Fatal/Serious Injury Chance', ax=axes[0])
axes[0].set_title('Mean Injury Chance by Make (Small Aircraft)')
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(data=large_15, x='Make', y='Fatal/Serious Injury Chance', ax=axes[1])
axes[1].set_title('Mean Injury Chance by Make (Large Aircraft)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Distribution of injury rates: small makes**

Use a violinplot to look at the distribution of the fraction of passengers serious/fatally injured for small airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
small_10 = small_plane_chance.sort_values().head(10).reset_index()

list(small_10['Make'])

filtered_small = small_plane[small_plane['Make'].isin(list(small_10['Make']))]
```

```python
# Create a violin plot for the 10 lowest Fatal/Serious Injury Chance values
plt.figure(figsize=(10, 6))
sns.violinplot(data=filtered_small, x='Make', y='Fatal/Serious Injury Chance')

# Customize the plot (optional)
plt.title('Distribution of Fatal/Serious Injury Chance for Small Aircraft')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

**Distribution of injury rates: large makes**

Use a stripplot to look at the distribution of the fraction of passengers serious/fatally injured for large airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
large_10 = large_plane_chance.sort_values().head(10).reset_index()

filtered_large = large_plane[large_plane['Make'].isin(list(large_10['Make']))]
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=filtered_large, x='Make', y='Fatal/Serious Injury Chance')

plt.title('Distribution of Serious/Fatal Injury Rates for Large Aircraft')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Evaluate the rate of aircraft destruction for both small and large aircraft by Make.** 

Sort your results and keep the lowest 15.

```python
small_plane_dest = small_plane.groupby('Make')['Destroyed?'].sum()/small_plane.groupby('Make')['Destroyed?'].count()
large_plane_dest = large_plane.groupby('Make')['Destroyed?'].sum()/large_plane.groupby('Make')['Destroyed?'].count()

sort_small_dest = small_plane_dest.sort_values().head(15).reset_index()
sort_large_dest = large_plane_dest.sort_values().head(15).reset_index()

sort_small_dest.columns = ['Make', 'Chance of Destruction']
sort_small_dest['Type'] = 'Small'

sort_large_dest.columns = ['Make', 'Chance of Destruction']
sort_large_dest['Type'] = 'Large'
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(data=sort_small_dest, x='Make', y='Chance of Destruction', ax=axes[0])
axes[0].set_title('Chance of Destruction by Make (Small Aircraft)')
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(data=sort_large_dest, x='Make', y='Chance of Destruction', ax=axes[1])
axes[1].set_title('Chance of Destruction by Make (Large Aircraft)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
small_merge = small_15.merge(sort_small_dest, on='Make')

small_merge.drop(columns = ['Type_x', 'Type_y'], inplace=True)

small_merge['Rec Value'] = small_merge['Fatal/Serious Injury Chance'] + small_merge['Chance of Destruction']

small_merge.sort_values('Rec Value')
```

```python
'''Recommendations can be provided based on makes that have both the lowest rates of destruction and lowest serious injury/fatal risk. The makes
listed above are the safest aircraft designers based on these parameters. 

For small planes, the safest makes would be Waco, Maule, and Let. For large planes, Mcdonnell Douglas, Boeing, and Embraer could generally be considered
the safest. 

Comments:
Fatal/Injury chance is a flawed metric inherently. A plane with many minor problems that never result in serious harm would appear safer than a plane
that flew flawlessly on many trips but crashed once due to weather or another uncontrolled variable. This could still result in major loss for the 
insurer due to needed replacement parts, etc. Similarly, total destructrion of a plane as a binary metric neglects many scenarios (such as substantial
damage) where the insurer would be liable. Such a large standard deviation and the violin and strip plots show the massive variation and unpredictablity
in accident outcomes. 
'''
```

### Analyze plane types
- plot the mean fatal/seriously injured fraction for both small and larger planes 
- also provide a distributional plot of your choice for the fatal/seriously injured fraction by airplane type (stripplot, violin, etc)  
- filter ensuring that you have at least ten individual examples in each model/make to average over

**Larger planes**

```python
#Count large planes by type
large_counts_type = large_plane['Type'].value_counts()

#filter out rows where there are less than 10 of that specific type
filtered_large_type = large_plane[large_plane['Type'].isin(large_counts_type[large_counts_type >= 10].index)]

#Group by type and get the mean on Fatal/Serious Injury Chance
large_10_type = filtered_large_type.groupby('Type')['Fatal/Serious Injury Chance'].mean()

#Get list of types with the lowest 10 means
large_10_list = large_10_type.sort_values().head(10).reset_index()

#Filter out types not in above list

filtered_large_10 = filtered_large_type[filtered_large_type['Type'].isin(list(large_10_list['Type']))]
```

**Smaller planes**
- for smaller planes, limit your plotted results to the makes with the 10 lowest mean serious/fatal injury fractions

```python
#Count small planes by type
small_counts_type = small_plane['Type'].value_counts()

#filter out rows where there are less than 10 of that specific type
filtered_small_type = small_plane[small_plane['Type'].isin(small_counts_type[small_counts_type >= 10].index)]

#Group by type and get the mean on Fatal/Serious Injury Chance
small_10_type = filtered_small_type.groupby('Type')['Fatal/Serious Injury Chance'].mean()

#Get list of types with the lowest 10 means
small_10_list = small_10_type.sort_values().head(10).reset_index()

#Filter out types not in above list

filtered_small_10 = filtered_small_type[filtered_small_type['Type'].isin(list(small_10_list['Type']))]
```

```python
small_graph = small_10_type.sort_values().head(10).reset_index()
large_graph = large_10_type.sort_values().reset_index()

small_graph.columns = ['Type', 'Fatal/Serious Injury Chance']

large_graph.columns = ['Type', 'Fatal/Serious Injury Chance']

small_graph.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(data=small_graph, x='Type', y='Fatal/Serious Injury Chance', ax=axes[0])
axes[0].set_title('Fatal/Serious Injury Chance by Type (Small Aircraft)')
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(data=large_graph, x='Type', y='Fatal/Serious Injury Chance', ax=axes[1])
axes[1].set_title('Fatal/Serious Injury Chance by Type (Large Aircraft)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.violinplot(data=filtered_large_10, x='Type', y='Fatal/Serious Injury Chance', ax=axes[0])
axes[0].set_title('Injury Chance by Type (Large Aircraft)')
axes[0].tick_params(axis='x', rotation=45)

sns.violinplot(data=filtered_small_10, x='Type', y='Fatal/Serious Injury Chance', ax=axes[1])
axes[1].set_title('Injury Chance by Type (Small Aircraft)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
